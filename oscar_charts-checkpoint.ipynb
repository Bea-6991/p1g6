```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import scipy.stats as st
from scipy.stats import linregress
from pprint import pprint

#setting up the axes for subplots
fig,axes=plt.subplots(2,2)
```

```python
#read in csv with clean data
output_data_file = "data_files/good_merged_data.csv"

with open(output_data_file, "r", encoding = "utf-8") as read_file:
    movies_df = pd.read_csv(read_file)
    
movies_df
```

```python
#setting up variables for each column
rating = movies_df['imdbRating']
votes = movies_df['imdbVotes']
revenue = movies_df['Revenue']
rated = movies_df['Rated']
roi = movies_df['ROI%']
budget = movies_df['Budget']
```

```python
#imdb Rating vs ROI Scatter plot
plt.scatter(rating, roi, edgecolors="black")
plt.xlabel("IMDB Rating")
plt.ylabel("ROI (%)")
plt.title("IMDB Rating vs ROI (%)")
plt.ylim(-50, 200)
plt.grid(b = True, axis = "both", which = "major")
plt.savefig("figures/imdbRating_vs_ROI.png")
plt.show
```

```python
#imdb Rating vs ROI without limits
plt.scatter(rating, roi, edgecolors="black")
plt.xlabel("IMDB Rating")
plt.ylabel("ROI (%)")
plt.title("IMDB Rating vs ROI (%)")
#plt.ylim(-350, 4000)
plt.grid(b = True, axis = "both", which = "major")
plt.savefig("figures/imdb_rating_vs_ROI_withoutlimit.png")
plt.show
```

```python
#imdb Rating vs Revenue
plt.scatter(rating, revenue, edgecolors="black")
plt.xlabel("IMDB Rating")
plt.ylabel("Revenue")
plt.title("IMDB Rating vs Revenue")
#plt.ylim(-350, 4000)
plt.grid(b = True, axis = "both", which = "major")
plt.savefig("figures/imdb_rating_vs_revenue.png")
plt.show
```

```python
#KEEP
#imdb_rating_vs_Revenue_vs_Budget
plt.scatter(rating, revenue, s = budget/1000000, edgecolor = 'black', alpha = 0.6)
plt.xlabel("IMDB Rating")
plt.ylabel("Revenue")
plt.title("IMDB Rating vs Revenue vs Budget")
plt.ticklabel_format(axis = 'y', style = 'plain')
#plt.ylim(-350, 4000)
#plt.grid(b = True, axis = "both", which = "major")
plt.savefig("figures/imdb_rating_vs_Revenue_vs_Budget.png")
plt.show
```

```python
#read in csv with clean production company
output_data_file = "data_files/good_merged_data.csv"

with open(output_data_file, "r", encoding = "utf-8") as read_file:
    production_df = pd.read_csv(read_file)

production_df
```

```python
#filtering out by top production companies
top_producers = ['Warner Bros. Pictures', 'Universal Pictures', '20th Century Fox', 'Sony Pictures', 
                 'Paramount Pictures', 'Lionsgate', 'New Line Cinema', 'Buena Vista', 'Fox Searchlight Pictures', 
                 'MGM', 'Miramax Films', 'Focus Features', 'Columbia Pictures', 'The Weinstein Company', 
                 'Walt Disney Studios']
top_producers_df = production_df.loc[production_df['Production'].isin(top_producers)]
top_producers_df
```

```python
#group by for each production company
prod_avg_df = top_producers_df.groupby('Production').mean()
prod_avg_df = prod_avg_df.sort_values(by = ['Revenue'], ascending = False)
pd.options.display.float_format = '{:,.10f}'.format
prod_avg_df
```

```python
#production company revenue and roi analysis
fig = plt.figure(figsize=[20, 10])
gs = plt.GridSpec(2, 3, figure=fig)

axes[0,0] = fig.add_subplot(gs[0, 0])
axes[0,1] = fig.add_subplot(gs[0, 1])

prod_avg_df.plot(y='Revenue',kind='bar',ax=axes[0,0],subplots=True)
prod_avg_df.plot(y='ROI%',kind='bar',ax=axes[0,1],subplots=True, color = 'r')


```

```python
#average revenue per production company
plt.bar(prod_avg_df.index, prod_avg_df['Revenue'])
plt.title("Average Revenue per Production Company")
plt.xlabel("Rating")
plt.ylabel("Average Revenue")
plt.xticks(rotation = 'vertical')
plt.ticklabel_format(axis = 'y', style = 'plain')
plt.tight_layout()
plt.savefig("figures/average_revenue_per_production_company.png")
plt.show()
```
